#Question 1: Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.

Answer:

SQL commands are divided into different categories based on the operations they perform on a database.

1. DDL (Data Definition Language):

DDL commands are used to create, modify, or delete database objects such as tables, databases, and indexes.
These commands change the structure of the database.

Example:

```python
CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    Name VARCHAR(50)
);
```

2. DML (Data Manipulation Language):

DML commands are used to insert, update, and delete data stored in database tables.
They work on the data rather than the database structure.

Example:

```python
INSERT INTO Students (StudentID, Name)
VALUES (1, 'Rahman');
```

3. DQL (Data Query Language):

DQL is used to retrieve data from one or more tables.
The main command in DQL is SELECT.

Example:

```python
SELECT * FROM Students;
```

Difference Summary

DDL	DML	DQL
Defines database structure	Manipulates data	Retrieves data
Uses CREATE, ALTER, DROP	Uses INSERT, UPDATE, DELETE	Uses SELECT
Changes schema	Changes records	Reads records

#Question 2: What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.

Answer:

SQL constraints are rules applied to table columns to ensure the accuracy, consistency, and integrity of the data stored in a database.

1. PRIMARY KEY

Ensures that each record has a unique value.
It does not allow duplicate or NULL values.

Scenario:
Each student in a college should have a unique StudentID.

Example:

```python
StudentID INT PRIMARY KEY
```

2. NOT NULL

Ensures that a column cannot have NULL values.
Every record must contain a value.

Scenario:
Every employee must have a name.

Example:

```python
EmployeeName VARCHAR(100) NOT NULL
```

3. FOREIGN KEY

Creates a relationship between two tables.
Ensures that values exist in the referenced table.

Scenario:
Every order must belong to an existing customer.

Example:

```python
CustomerID INT,
FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
```

Benefits of Constraints

Prevents invalid data.
Maintains data integrity.
Reduces duplication.
Improves database reliability.

#Question 3: Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?

Answer:

The LIMIT clause specifies the maximum number of rows to return from a query.

The OFFSET clause specifies how many rows should be skipped before returning the results.

Difference

LIMIT	OFFSET
Restricts the number of rows returned	Skips a specified number of rows
Controls page size	Controls page position

For pagination:

Page size = 10 records
Third page means skipping the first 20 records.

SQL Query:

```python
SELECT *
FROM Products
LIMIT 10 OFFSET 20;
```

Explanation:

LIMIT 10 returns 10 records.
OFFSET 20 skips the first 20 records and starts displaying from the 21st record.

#Question 4: What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.

Answer:

A Common Table Expression (CTE) is a temporary named result set created using the WITH keyword. It exists only during the execution of a single SQL query.

Benefits of CTE

Makes SQL queries easier to read.
Simplifies complex queries.
Improves code organization.
Can be reused within the same query.
Supports recursive queries.

Example:

```python
WITH HighSalaryEmployees AS
(
    SELECT EmployeeID, EmployeeName, Salary
    FROM Employees
    WHERE Salary > 50000
)
SELECT *
FROM HighSalaryEmployees;
```

Explanation:
The CTE first stores employees with salaries greater than 50,000. The main query then retrieves the data from the CTE.

#Question 5: Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).

Answer:

Normalization is the process of organizing data in a database to reduce redundancy and improve data integrity.

Primary Goals
Eliminate duplicate data.
Maintain consistency.
Improve storage efficiency.
Simplify database maintenance.
Prevent update, insert, and delete anomalies.
First Normal Form (1NF)

A table is in 1NF if:

Each column contains only one value.
There are no repeating groups.
Every row is unique.

Example:
Instead of storing multiple phone numbers in one column, each phone number should be stored separately.

Second Normal Form (2NF)

A table is in 2NF if:

It is already in 1NF.
Every non-key attribute depends on the entire primary key.

This removes partial dependency.

Third Normal Form (3NF)

A table is in 3NF if:

It is already in 2NF.
Non-key attributes depend only on the primary key and not on other non-key attributes.

This removes transitive dependency.

Example

Instead of storing:

StudentID	StudentName	DepartmentName	HOD

Create separate tables:

Students

| StudentID | StudentName | DepartmentID |

Departments

| DepartmentID | DepartmentName | HOD |

This avoids duplication and makes updates easier.

#Question 7
Customer Report with Total Orders
SELECT
    c.CustomerName,
    c.Email,
    COUNT(o.OrderID) AS TotalNumberOfOrders
FROM Customers c
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID
GROUP BY
    c.CustomerID,
    c.CustomerName,
    c.Email
ORDER BY c.CustomerName;

#Question 9
Top 2 Most Expensive Products in Each Category using CTE and ROW_NUMBER()
WITH RankedProducts AS
(
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER
        (
            PARTITION BY c.CategoryName
            ORDER BY p.Price DESC
        ) AS RowNum
    FROM Products p
    JOIN Categories c
    ON p.CategoryID = c.CategoryID
)

SELECT
    CategoryName,
    ProductName,
    Price
FROM RankedProducts
WHERE RowNum <= 2;
Question 10 (Sakila Database)
1. Top 5 Customers by Amount Spent
SELECT
    CONCAT(c.first_name,' ',c.last_name) AS CustomerName,
    c.email,
    SUM(p.amount) AS TotalAmountSpent
FROM customer c
JOIN payment p
ON c.customer_id=p.customer_id
GROUP BY
    c.customer_id,
    CustomerName,
    c.email
ORDER BY TotalAmountSpent DESC
LIMIT 5;
2. Top 3 Movie Categories by Rental Count
SELECT
    cat.name AS CategoryName,
    COUNT(*) AS RentalCount
FROM rental r
JOIN inventory i
ON r.inventory_id=i.inventory_id
JOIN film_category fc
ON i.film_id=fc.film_id
JOIN category cat
ON fc.category_id=cat.category_id
GROUP BY cat.name
ORDER BY RentalCount DESC
LIMIT 3;
3. Films Available at Each Store and Never Rented
SELECT
    s.store_id,
    COUNT(i.inventory_id) AS TotalFilms,
    SUM(
        CASE
            WHEN r.rental_id IS NULL THEN 1
            ELSE 0
        END
    ) AS NeverRented
FROM store s
JOIN inventory i
ON s.store_id=i.store_id
LEFT JOIN rental r
ON i.inventory_id=r.inventory_id
GROUP BY s.store_id;
4. Monthly Revenue for 2023
SELECT
    MONTH(payment_date) AS Month,
    SUM(amount) AS Revenue
FROM payment
WHERE YEAR(payment_date)=2023
GROUP BY MONTH(payment_date)
ORDER BY Month;
5. Customers Who Rented More Than 10 Times in the Last 6 Months
SELECT
    CONCAT(c.first_name,' ',c.last_name) AS CustomerName,
    COUNT(r.rental_id) AS TotalRentals
FROM customer c
JOIN rental r
ON c.customer_id=r.customer_id
WHERE r.rental_date >= DATE_SUB(CURDATE(),INTERVAL 6 MONTH)
GROUP BY
    c.customer_id,
    CustomerName
HAVING COUNT(r.rental_id)>10
ORDER BY TotalRentals DESC;
